```python
# To use objects and functions of pandas library
import pandas as pd
```

```python
#read csv file and store values to the variable
dataset = pd.read_csv("50_Comp_Profit_Analysis.csv")
```

```python
dataset
```

```python
dataset.columns
```

```python
#nominal data convert 0 and 1
#astype(int) method to convert the boolean values to integers
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
#split input and output
independent = dataset[['R & D', 'COSTING', 'MARKETING', 'STATE_Coimbatore',
       'STATE_Delhi', 'STATE_Karur', 'STATE_Madurai', 'STATE_Mumbai',
       'STATE_Trichy']]
dependent = dataset[['PROFIT']]
```

```python
independent
```

```python
dependent
```

```python
#split training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent, test_size=0.30,random_state=0)
```

```python
#create model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
weight = regressor.coef_
weight
```

```python
bias = regressor. intercept_
bias
```

```python
# to predict the dependent variable and value store in variable
Y_predict = regressor.predict(X_test)
#calculate the R-squared score
from sklearn.metrics import r2_score
#To indicates a better fit or not.
r_score = r2_score(Y_test,Y_predict)
```

```python
r_score
```

```python
#defines a filename for saving purpose
import pickle
#This line defines a filename for saving the serialized model .sav commonly used for saved models
filename = "Finalized_Multiple_Linear_Model.sav"
```

```python
#To save the trained linear regression model
pickle.dump(regressor, open(filename, 'wb'))
```

```python
#The loaded model to make predictions, opens the file and read as binary file
model = pickle.load(open("Finalized_Multiple_Linear_Model.sav",'rb'))
```
